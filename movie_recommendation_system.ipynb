```python
df=pd.read_csv('movies_metadata.csv', low_memory=False)
df=df[['title','overview','genres','tagline','vote_average','popularity']]
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['title'])
```

```python
df['overview']=df['overview'].fillna('')
df['tagline']=df['tagline'].fillna('')
```

```python
df.iloc[0]['genres']
```

```python
import ast
df['genres']=df['genres'].apply(lambda x: " ".join([i['name'] for i in ast.literal_eval(x)]))
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['tags']=df['overview']+" "+df['tagline']+" "+df['genres']
```

```python
df['tags'][0]
```

```python
df=df.drop(columns=['overview','tagline','genres'])
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stopwords=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
def preprocess_text(text):
  text=str(text.lower()) #lowercasing the text
  text=re.sub(r'[^\w\s]','',text) #removing the punctuation
  words=text.split()
  #remove stopwords
  words=[word for word in words if word not in stopwords]
  #lemmatizer
  words=[lemmatizer.lemmatize(word) for word in words]
  return " ".join(words)
```

```python
df['tags']=df['tags'].apply(preprocess_text)
```

```python
df['tags'][0]
```

```python
df=df.reset_index(drop=True)
```

```python
df.drop_duplicates(subset=['title'],inplace=True)
indices=pd.Series(df.index,index=df['title'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix=tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title,n=10):
  if title not in indices:
    return ['Movie not found']

  idx=indices[title]
  sim_score=cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
  similar_idx=sim_score.argsort()[::-1][1:n+1]
  return df['title'].iloc[similar_idx]
```

```python
recommend('Avatar')
```

```python
import pickle
pickle.dump(tfidf_matrix,open('tfidf_matrix.pkl','wb'))
pickle.dump(indices,open('indices.pkl','wb'))
df.to_pickle('df.pkl')
pickle.dump(tfidf,open('tfidf.pkl','wb'))
```
